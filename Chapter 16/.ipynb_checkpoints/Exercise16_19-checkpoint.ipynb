```python
from emnist import extract_training_samples, extract_test_samples
```

```python
X_train, y_train = extract_training_samples('byclass')
X_test, y_test = extract_test_samples('byclass')
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=2)
```

```python
indices = np.random.choice(np.arange(1000), 30, replace=False )

figure, axes = plt.subplots(nrows=5, ncols=6, figsize=(16, 9))

for item in zip(axes.ravel(), X_train[indices], y_train[indices]):
    axes, image, target = item
    axes.imshow(image, cmap=plt.cm.gray_r)
    axes.set_xticks([])
    axes.set_yticks([])
    axes.set_title(target)
plt.tight_layout()
```

```python
X_train = X_train.reshape(697932, 28, 28, 1)
X_test = X_test.reshape(116323, 28, 28, 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train = X_train.astype('float32') / 255
X_test = X_test.astype('float32') / 255
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train = to_categorical(y_train)
```

```python
y_train[9]
```

```python
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Conv2D, Dense, MaxPooling2D, Flatten, Dropout
```

```python
cnn = Sequential()
```

```python
cnn.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
```

```python
cnn.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
cnn.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
```

```python
cnn.add(MaxPooling2D(pool_size=(2, 2)))
```

```python
cnn.add(Flatten())
```

```python
cnn.add(Dense(units=128, activation='relu'))
```

```python
cnn.add(Dense(units=62, activation='softmax'))
```

```python
cnn.summary()
```

```python
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, batch_size=64, epochs=5, validation_split=0.1)
```

```python
loss, accuracy = cnn.evaluate(X_test, y_test)
```
